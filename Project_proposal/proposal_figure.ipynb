```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
from numpy.random import rand
```

```python
with plt.xkcd():
    # Based on "Stove Ownership" from XKCD by Randall Monroe
    # http://xkcd.com/418/

    fig = plt.figure()
    ax = fig.add_axes((0.1, 0.2, 0.8, 0.7))
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    plt.xticks([])
    plt.yticks([])

    x, y = rand(2,20)*0.5
    plt.scatter(x, y)
    x, y = rand(2,20)*0.5 + 0.5
    plt.scatter(x, y)
    
    plt.xlabel('IESE Cities in Motion Index')
    plt.ylabel('Gentrification Index')
    plt.title('Smart Cities are gentrified')
    ax.annotate('smart & gentrified', xy=(0.5, 0.8), xytext=(0., 0.8), 
                arrowprops=dict(facecolor='black', shrink=0.05))
    ax.annotate('not smart & not gentrified', xy=(0.4, 0.2), xytext=(0.5, 0.2), 
                arrowprops=dict(facecolor='black', shrink=0.05))
    
    plt.savefig('postive_corr.png')
```

```python
with plt.xkcd():
    # Based on "Stove Ownership" from XKCD by Randall Monroe
    # http://xkcd.com/418/

    fig = plt.figure()
    ax = fig.add_axes((0.1, 0.2, 0.8, 0.7))
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    plt.xticks([])
    plt.yticks([])

    x = rand(1,20)*0.5
    y = rand(1,20)*0.5 + 0.5
    plt.scatter(x, y)
    x = rand(1,20)*0.5 + 0.5
    y = rand(1,20)*0.5
    plt.scatter(x, y)
    
    plt.xlabel('IESE Cities in Motion Index')
    plt.ylabel('Gentrification Index')
    plt.title('Smart Cities are not gentrified')
    ax.annotate('not smart & gentrified', xy=(0.5, 0.8), xytext=(0.6, 0.8), 
                arrowprops=dict(facecolor='black', shrink=0.05))
    ax.annotate('smart & not gentrified', xy=(0.57, 0.2), xytext=(0., 0.2), 
                arrowprops=dict(facecolor='black', shrink=0.05))
    
    plt.savefig('negative_corr.png')
```
