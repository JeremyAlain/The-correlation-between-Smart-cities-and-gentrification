#### This notebook is here to regress over our incomes. We namely want that our average houshold incomes for each district spans from 2008 - 2016 

```python
import numpy as np
import pandas as pd
```

# Smart Cities 

# Hamburg

```python
sheet = pd.read_excel("Cities Household Income per district/Smart-Cities houshold income per district.xlsx","Hamburg")
hamburg = sheet.as_matrix()
regressed_income = np.zeros((hamburg.shape[0],9))
dates = [2004, 2007, 2010]

# Fill in what we already know 
#2010
regressed_income[:,2] = hamburg[:,2]

# Calculate not known years(2008,2009,2011-2016)
for pc in range(hamburg.shape[0]):
    reg_fit = np.polyfit(dates, hamburg[pc,:], deg= 1)
    poly_fit = np.poly1d(reg_fit)
    
    # Fill in the regressed incomes 
    # 2008
    regressed_income[pc,0] = poly_fit(2008)
    #2009
    regressed_income[pc,1] = poly_fit(2009)
    #...
    regressed_income[pc,3] = poly_fit(2011)
    regressed_income[pc,4] = poly_fit(2012)
    regressed_income[pc,5] = poly_fit(2013)
    regressed_income[pc,6] = poly_fit(2014)
    regressed_income[pc,7] = poly_fit(2015)
    regressed_income[pc,8] = poly_fit(2016)
print(regressed_income)
```

```python
writer = pd.ExcelWriter('Cities Household Income per district/Regressed_smart_city_household_income.xlsx', engine='xlsxwriter')
df_hamburg = pd.DataFrame(regressed_income)
df_hamburg.to_excel(writer, sheet_name='Hamburg')
```
