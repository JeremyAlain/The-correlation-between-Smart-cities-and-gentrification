```python
from bs4 import BeautifulSoup
import urllib.request as urlib
import pandas as pd
import numpy as np
```

# Sydney

```python
# define all postal codes 
input_file = pd.read_csv('sydney_postalcodes.csv', header=None)
postalcode = input_file.as_matrix()
print(postalcode.shape)
# define years 
years = ["2006", "2011", "2016"]
database = []
```

```python
i = 0
for year in years: 
    print("Year:" , year)
    database.append([])
    i+=1
    for pc in postalcode: 
        pc = pc[0]
        site = "http://www.censusdata.abs.gov.au/census_services/getproduct/census/"+year+"/quickstat/POA"+str(pc)
        page = urlib.urlopen(site)
        soup = BeautifulSoup(page)
        if(year != "2016"):
            table = soup.find_all('table', class_='summaryTable')
            for table in table: 
                tr = table.find_all('tr')
                for row in tr:
                    cells = row.find_all('td')
                    if("Median incomes" in str(cells[0])):
                        c = cells[1].string
                        c = c[1:]
                        c = c.replace(',', '')
                        nr = int(c)
                        database[i-1].append(nr)
                        print("    PC ", pc)
        else: 
            cells = []
            table = soup.find_all('table', class_='summaryTable')
            for table in table: 
                tr = table.find_all('tr')
                for row in tr:
                    cell = row.find_all('td')
                    cells.append(cell[0])
            if(len(cells) != 1):
                c = cells[10].string
                c = c[1:]
                c = c.replace(',', '')
                nr = int(c)
                database[i-1].append(nr)
                print("pc", pc)
            else:
                print("pc: ", pc, "Error")
```

```python
print(database)
print(len(database[0]))
print(len(database[1]))
print(len(database[2]))
```

```python
bla = database
bla2 = database
```

```python
print(database)
```

```python
for i in range(3):
    while(len(database[i]) <= 198):
        database[i].append(0)
    print(len(database[i]))
db = np.zeros((199,3), np.int32)
db[:,0] = database[0]
db[:, 1] = database[1]
db[:, 2] = database[2]
```

```python
print(db[0])
df = pd.DataFrame(db)
df.to_csv("sydney_income.csv")
```
